## Let analyse the from smoking point of view 

```python
smokers=df[df["smoking"]==1]
smokers.head()
smokers_died=smokers[smokers["DEATH_EVENT"]==1]
smokers_died.head()
smokers_survived=smokers[smokers["DEATH_EVENT"]==0]
smokers_survived.head()

non_smokers=df[df["smoking"]==0]
non_smokers.head()
non_smokers_died=non_smokers[non_smokers["DEATH_EVENT"]==1]
non_smokers_survived=non_smokers[non_smokers["DEATH_EVENT"]==0]
```

```python
smoke_cat=["smokers_died","smokers_survived","non_smokers_died","non_smokers_survived"]
smoke_val=[len(smokers_died),len(smokers_survived),len(non_smokers_died),len(non_smokers_survived)]
```

```python
plt.figure(figsize=(8,8))
plt.pie(smoke_val,labels=smoke_cat,autopct="%1.1f%%")
plt.legend()
plt.title("Smoker_Died_Survived")
circle = plt.Circle((0,0), 0.8, color="black")
p = plt.gcf()
p.gca().add_artist(circle)
```

## Data Modeling and Feature Scaling  and Fitting into Machine Learning Algorithms 

```python
df_continious.head()
```

```python
x=df_continious
y=df["DEATH_EVENT"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

## Scaling train Data 
standard Scaler

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
print(x_train_scaled,x_test_scaled)
```

## Inplementing Machine Learning Algorithms for Predictions 

## 1 Logistic Rigression 


```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression()
```

```python
lr_model.fit(x_train_scaled,y_train)
lr_prediction=lr_model.predict(x_test_scaled)
y_predicted=lr_prediction
```

```python
lr_model_accuracy=accuracy_score(y_predicted,y_test)
```

```python
lr_model_accuracy* 100
```

```python
accuracy_list=[]   # list to hold the accuracy of all algorithms 
```

```python
accuracy_list.append(lr_model_accuracy* 100)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,y_predicted)
```

```python
ConfusionMatrixDisplay(confusion_matrix=cm)
cm
```

## 2. Support Vector Machine 

```python
from sklearn.svm import SVC
```

```python
model_svm= SVC()
```

```python
model_svm.fit(x_train_scaled,y_train)
y_predicted=model_svm.predict(x_test_scaled)
svm_accuracy=accuracy_score(y_predicted,y_test)
svm_accuracy
```

```python
accuracy_list.append(svm_accuracy*100)
```

## 3.NaiveBayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_nb=GaussianNB()
```

```python
model_nb.fit(x_train_scaled,y_train)
y_predicted=model_nb.predict(x_test_scaled)
accuracy_nb=accuracy_score(y_predicted,y_test)
```

```python
accuracy_nb*100
```

```python
accuracy_list.append(accuracy_nb*100)
```

## 4. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DCT=DecisionTreeClassifier()
model_DCT.fit(x_train_scaled,y_train)
y_predicted=model_DCT.predict(x_test_scaled)
accuracy_DCT=accuracy_score(y_predicted,y_test)
```

```python
accuracy_DCT
```

```python
accuracy_list.append(accuracy_DCT*100)
```

```python
accuracy_list
```

```python
accuracy_list_update=[]

for i in accuracy_list:
    accuracy_list_update.append(round(i,2))

print(accuracy_list_update)    
    
```

```python
ml_methods=["LogisticRegression","SupportVectorMachine","NaiveBayes","DecisionTreeClassifier"]
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=ml_methods,y=accuracy_list_update,hue=ml_methods)
plt.xlabel("Classifiers")
plt.ylabel("Accuracy")
plt.title("Accuracy Achieved by different classifiers")
```
